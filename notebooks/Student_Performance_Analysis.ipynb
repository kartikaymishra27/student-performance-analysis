# Student Performance Analysis

## Problem Statement
The objective of this project is to analyze how study hours and attendance
affect the academic performance of students.


```python
import pandas as pd

data = {
    'study_hours': [1, 2, 3, 4, 5, 6, 7, 8],
    'attendance': [55, 60, 65, 70, 75, 80, 85, 90],
    'math_score': [40, 45, 55, 60, 70, 78, 85, 92],
    'science_score': [42, 48, 58, 62, 72, 80, 88, 90],
    'english_score': [44, 50, 60, 65, 74, 82, 86, 91]
}

df = pd.DataFrame(data)
df
```

## Score Calculation
Total and average scores were calculated for each student to evaluate
overall academic performance across subjects.


```python
df['total_score'] = (
    df['math_score'] +
    df['science_score'] +
    df['english_score']
)

df
```

```python
df['average_score'] = df['total_score'] / 3
df
```

## Performance Analysis
The student with the highest average score represents the top performer,
while the student with the lowest average score indicates areas where
additional academic support may be required.


## Visualization: Study Hours vs Performance
This graph shows the relationship between study hours and average score.


```python
import matplotlib.pyplot as plt

plt.scatter(df['study_hours'], df['average_score'])
plt.xlabel("Study Hours")
plt.ylabel("Average Score")
plt.title("Study Hours vs Average Score")
plt.show()
```

## Graph Interpretation
The scatter plot indicates a positive relationship between study hours
and average score. Students who study more hours tend to perform better.


## Conclusion
This analysis shows that students who study more hours tend to achieve
higher average scores. The results indicate a positive relationship
between study habits and academic performance. Consistent study and
attendance play a crucial role in improving results.
